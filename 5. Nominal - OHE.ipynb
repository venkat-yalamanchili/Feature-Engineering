```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe
df = pd.DataFrame({
    'color' : ['red','blue','green','green','red','green','blue','green']
})
```

```python
df.head()
```

```python
## Create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

This line of code is a one-step command to both "learn" the categories in your data and then convert them into numbers.

1. encoder.fit_transform(...)
This method combines two distinct steps that scikit-learn transformers perform:

fit: The encoder looks at the color column and identifies all unique categories (e.g., "Red", "Blue", "Green"). It builds a mapping (e.g., Red = 0, Blue = 1).

transform: It applies that mapping to the data, replacing the text strings with the corresponding numerical values.

By using fit_transform, you do both at once, which is more efficient during the initial training phase of your model.

```python
# perform fit and transform
encoder.fit_transform(df[['color']])
```

```python
encoder.fit_transform(df[['color']]).toarray()
```

Why red got(0,0,1) not (1,0,0) because it is done alphabetically so blue got (1,0,0)

```python
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encode_df = pd.DataFrame(encoded)
```

```python
encode_df
```

```python
encode_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encode_df
```

```python
pd.concat([df,encode_df],axis=1)
```

## Note
2. Why [[]] instead of []?
This is a common point of confusion in Pandas. The difference comes down to the shape of the object being returned.

Single Brackets: df['color']
Returns: A Pandas Series.

Shape: 1-Dimensional.

Analogy: A single list of items.

Why it fails here: Scikit-learn encoders (like OneHotEncoder or OrdinalEncoder) are designed to handle multiple features at once. They strictly expect a 2D array (rows and columns), even if you are only passing in one column.

Double Brackets: df[['color']]
Returns: A Pandas DataFrame.

Shape: 2-Dimensional.

Analogy: A table that happens to have only one column.

Why we use it: This satisfies the "2D requirement" of the encoder. It tells Pandas, "Give me a subset of this table containing these columns," which keeps the result in a table (DataFrame) format.
