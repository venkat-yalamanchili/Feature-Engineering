5 Number Summary and box plot : Minimum,Q1,Median,Q3,Maximum
1. To find the ouliers we need to fing lower and higher fence values outside these fence are outliers(Go through Notes under 5 number summary for better understanding)

```python
import numpy as np
```

```python
lst_marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum,Q1,median,Q3,maximum = np.quantile(lst_marks,[0,0.25,0.5,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
```

```python
lower_fence,higher_fence
```

```python
import seaborn as sns
```

```python
sns.boxplot(lst_marks, color='cyan')
```
