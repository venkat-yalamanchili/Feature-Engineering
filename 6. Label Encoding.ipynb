### Label Encoding
label encoding and ordinal encoding are two techniques use to encode categorical data as numerical data (Go through Notes)

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    'color' : ['red','blue','green','green','red','green','blue','green']
})
```

```python
df.head()
```

```python
## Create instance of labelEncoder
le = LabelEncoder()
encoded_data = le.fit_transform(df['color'])  #df[] is series and d[[]] is dataframe
```

```python
encoded_data
```

```python
#We can do it for any new data point 
le.transform(['red','green'])
```

## Ordinal Encoding
you can manually assign the numbers to preserve the real world scale(Ex: low=0,medium=1,high=2)

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size':['small','large','medium','large','large','medium','small']
})
```

```python
df.head()
```

```python
## Create an instance of Ordinal encoder
oe = OrdinalEncoder(categories=[['small','medium','large']]) # numbering is given by this order
ordinal_data = oe.fit_transform(df[['size']])
```

```python
ordinal_data
```

You can see small has 0, medium has 1 and large has 2
